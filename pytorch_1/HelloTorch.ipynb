#### 1.1.5 运算的内存开销

每个操作都会新开内存来存储运算结果

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

把特定结果放到特定内存，torch.zeros_like()方法可以创建形状一致元素为0的张量矩阵
在张量矩阵参数Z后加上[:],后面加上想要写进Z对应内存里的参数。这是开辟了临时内存存储计算结果。

```python
Z = torch.zeros_like(Y)
before = id(Z)
Z[:] = X + Y
id(Z) == before
```

运算符全名函数中的out函数可以实现不开辟临时内存存储结果

```python
torch.add(X,Y,out=Z)
id(Z) == before
```
