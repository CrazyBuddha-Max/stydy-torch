# Use pytorch to do deeplearning

## 1. 预备知识

### 1.1数据操作

#### 1.1.1 创建Tensor

```python
import torch

print("pytorch version: {}".format(torch.__version__))
```

##### 构建一个行向量

```python
tensor_a = torch.arange(0,12) 
tensor_a
```

```python
print("张量的存储位置：{}".format(tensor_a.device))
```

##### shape属性可以获取tensor实例的形状

```python
tensor_a.shape
```

##### numel()方法可以获取tensor实例中元素的总数

```python
tensor_a.numel()
```

##### view()方法可以改变向量形状

```python
X = tensor_a.view(3,4)
X
```

##### torch.zero()和ones()方法可以生成固定值为0、1的张量矩阵

```python
torch.zeros((2,3,4))
```

```python
torch.ones((3,4))
```

##### python的list可以创建指定的张量矩阵

```python
Y = torch.tensor([[2,1,4,3],[1,2,3,4],[4,3,2,1]])
Y
```

##### torch.randn()方法可以创建一个指定形状，每个元素的值都随机采样于均值为0，标准差为1的正态分布的张量矩阵

```python
torch.randn((3,4))
```

#### 1.1.2 运算

##### 加法

```python
X + Y
```

##### 按元素乘法

```python
X * Y
```

##### 按元素除法

```python
X.float() / Y.float()
```

##### 按元素做指数运算

```python
torch.exp(Y.float())
```

##### 乘法

```python
X @ Y.t()
```

##### torch.cat()方法可以创建多个Tensor连结

```python
# 0为纵向拼接，1为横向拼接
torch.cat((X,Y),0),torch.cat((X,Y),1)
```

##### 判断两个张量矩阵是否相等

```python
X == Y
```

##### 对单个张量矩阵求和得到一个值

```python
torch.sum(X)
```

##### .item()方法可以把结果变换为Python中的标量

```python
torch.norm(X.float()).item()
```

#### 1.1.3 广播机制

```python
A = torch.arange(0,3).view(3,1)
B = torch.arange(0,2).view(1,2)

A,B
```

```python
A + B
```
